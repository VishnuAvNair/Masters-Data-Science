## Question 2:  Do republican electors have more power?

Power is defined by the Merriam-Webster (https://www.merriam-webster.com/dictionary/power) as the "ability to act or produce an effect".  **On average it took roughly 20% more Democrat electors to win a seat for a Democrate candidate, than it took Republican electors to elect a Republican candidate.**  In that sense, Republican electors had more power in the 2016 House of Representative elections.

```python
_ = get_dviz.plot_when_less_is_more_national(overall)
```

## Question 3:  How do state and national numbers compare?

Numbers at the state level reveal even more strickingly that Democrat electors needed more votes on average to elect their representatives to the House of Representatives than did Republican electors.  The horizontal lines on the graphs below show the average votes needed to elect a candidate of each party.  The problem we saw with Democratic electors needing to cast more votes to elect their candidates is exacerbated when we look at numbers at the state level.  It is true that Democrats had the upper hand in states like Oregon and Maryland in particular.  But Democrats were disadvantaged in many more states, and often in a more significant way. ## Question 5:  How many votes are being wasted?

** For districts with one uncontested candidate: <br> **
* all votes from winner beyond simple majority are considered wasted votes <br>

** For districts with two candidates: <br> **
* all votes from losing candidate are considered wasted votes <br>
* all votes from winner beyond simple majority are considered wasted votes <br>

** For districts with more than two candidates: <br> **
* all votes from losing candidates are considered wasted votes <br>
* all votes from winner beyond second most voted candidate plus 1, are considered wasted votes <br>

** We exclude states with a single district, since gerrymandering doesn't apply in those cases. **




```python
wasted = get_data.get_wasted_votes(election_results)
wasted.head()
```

```python
fig, ax = plt.subplots(figsize=(6, 10))
_ = get_dviz.plot_wasted_votes(wasted, fig, ax)
```
