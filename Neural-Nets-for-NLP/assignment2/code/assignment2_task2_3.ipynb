```python
# here we repeat processing described at https://web.stanford.edu/~jurafsky/ws97/manual.august1.html
#
# "There were 220 tags used in the coding; 130 of these occurred less than 10 times each, so for our initial 
# experiments we clustered the 220 tags into 42 larger classes. We did the clustering by removing the secondary 
# carat-dimensions (^2,^g,^m,^r,^e,^q,^d), with 5 exceptions. The exceptions: we left qy^d (Declarative yes-no 
# Questions) , qw^d (Declarative wh-questions) and b^m (Signal-Understanding-via-Mimic), and we folded the few 
# examples of nn^e into ng, and ny^e into na. Then, we grouped together some tags that had very little training 
# data; those tags that appear in the following list were grouped with other tags on the same line.

# qr qy
# fe ba
# oo co cc
# fx sv
# fo o fw " by bc
# aap am
# arp nd

# We also removed any line with a "@" (since @ marked slash-units with bad segmentation)."
```

```python
def get_tag_no(tag_in):
    if tag_in.find('@') != -1:   # bad segmentation; mark for deletion
        return -1
    elif tag_in[:2] == 'sd':
        return 0
    elif tag_in == 'b':
        return 1
    elif tag_in in ['fx', 'sv']:
        return 2
    elif tag_in[:2] == 'sv':
        return 2
    elif tag_in in ['aa']:
        return 3
    elif tag_in in ['fx', 'sv']:
        return 4
    elif tag_in in ['fe', 'ba']:
        return 5
    elif tag_in[:2] == 'ba':
        return 5
    elif tag_in in ['qr','qy']: 
        return 6
    elif tag_in[:1] == 'x': 
        return 7
    elif tag_in == 'ny': 
        return 8  
    elif tag_in[:2] == 'fc': 
        return 9 
    elif tag_in[0] == '%': 
        return 10 
    elif tag_in == 'qw': 
        return 11
    elif tag_in == 'nn': 
        return 12
    elif tag_in[:2] == 'bk': 
        return 13
    elif tag_in[:1] == 'h': 
        return 14
    elif tag_in == 'qy^d': 
        return 15
    elif tag_in[:2] == 'qy': 
        return 6
    elif tag_in in ['fo', 'o', 'fw', '"', 'by', 'bc']:
        return 16
    elif tag_in[:2] == 'fw':
        return 16
    elif tag_in[:2] == 'fo':
        return 16
    elif tag_in[:2] == 'bc':
        return 16
    elif tag_in[:2] == 'bh':
        return 17    
    elif tag_in[:2] == '^q':
        return 18    
    elif tag_in[:2] == 'bf':
        return 19    
    elif tag_in in ['na', 'ny^e']:
        return 20  
    elif tag_in[:2] == 'na':
        return 20
    elif tag_in[:2] == 'ny':
        return 20
    elif tag_in[:2] == 'ad':
        return 21    
    elif tag_in[:2] == '^2':
        return 22    
    elif tag_in[:3] == 'b^m':
        return 23    
    elif tag_in[:2] == 'qo':
        return 24    
    elif tag_in[:2] == 'qh':
        return 25    
    elif tag_in[:2] == '^h':
        return 26    
    elif tag_in == 'ar':
        return 27    
    elif tag_in in ['ng', 'nn^e']:
        return 28    
    elif tag_in[:2] == 'nn': 
        return 12
    elif tag_in[:2] == 'ng':
        return 28    
    elif tag_in[:2] == 'br':
        return 29    
    elif tag_in[:2] == 'no':
        return 30    
    elif tag_in[:2] == 'fp':
        return 31 
    elif tag_in[:3] == 'qrr':
        return 32
    elif tag_in[:2] == 'qr': 
        return 6
    elif tag_in in ['arp', 'nd']:
        return 33
    elif tag_in[:2] == 'nd':
        return 33
    elif tag_in[:2] == 'ar':
        return 27    
    elif tag_in == 't3':
        return 34
    elif tag_in[:2] == 'oo':
        return 35
    elif tag_in[:1] == 'o':
        return 16
    elif tag_in[:2] == 'co':
        return 35
    elif tag_in[:2] == 'cc':
        return 35
    elif tag_in[:2] == 't1':
        return 36
    elif tag_in[:2] == 'bd':
        return 37
    elif tag_in in ['aap', 'am']:
        return 38
    elif tag_in[:3] == 'aap':
        return 38
    elif tag_in[:2] == 'am':
        return 38
    elif tag_in[:2] == 'aa':
        return 3
    elif tag_in == '^g':
        return 39
    elif tag_in == 'qw^d':
        return 40
    elif tag_in[:2] == 'qw':
        return 11
    elif tag_in[:2] == 'fa':
        return 41
    elif tag_in[:2] == 'ft':
        return 42
    elif tag_in[:1] == 'b':
        return 1
    elif tag_in[:1] == '+':
        return 43
    else:
        return 99

trans_detail['tag_no'] = trans_detail['act_tag'].apply(lambda x: get_tag_no(x))
print('Original class count: {}'.format(len(trans_detail.act_tag.unique())))
# delete malformed rows, i.e., rows containing '@'
trans_detail_f = trans_detail[trans_detail['tag_no']!=-1].copy()
print('Malformed rows deleted: {}'.format(len(trans_detail[trans_detail['tag_no']==-1].index)))
print('Remaining rows: {}'.format(trans_detail_f.shape[0]))
print('Final class count: {}'.format(len(trans_detail_f.tag_no.unique())))
print('Unmap row count: {}'.format(len(trans_detail[trans_detail['tag_no']==99]['act_tag'].unique())))
```

## Enhanced MLP with BOW features extracted from transcript text - Prosodic features included

```python
%%time

# size of BOW
BOW_LEN = 100000

from sklearn.feature_extraction.text import TfidfVectorizer as tfidf
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split

vectorizer = tfidf( 
                    analyzer='char',      
                    token_pattern='.*',   # will consider all chars, to capture prosody info
                    ngram_range=(1,5),
                    max_features=BOW_LEN
                  )
X = vectorizer.fit_transform(trans_detail_f.text.values)

lb = LabelBinarizer()
y = lb.fit_transform(trans_detail_f.tag_no.values.astype('str'))

# will put aside 19/1,155 of the data for testing, which rounds up to 2%
# based on test partition size documented in "Hamed Khanpour, Nishitha Guntakandla, and Rodney Nielsen. 2016. 
# Dialogue Act Classification in Domain-Independent Conversations Using a Deep Recurrent Neural Network. In COLING."
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=round(19/1155,2), shuffle=True)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
%%time

import keras
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Input, Dense, Dropout
from keras.optimizers import SGD
from keras import regularizers
from keras.callbacks import EarlyStopping

EPOCHS = 100
CLASSES_NO = y.shape[1]
BATCH_SIZE = 128

def get_model(dim_len, num_classes):
    
    input = Input(shape=(dim_len,), sparse=True)
    dense = Dense(256, activation='relu')(input)
    drop  = Dropout(0.5)(dense)
    dense2 = Dense(256, activation='relu')(drop)
    drop2  = Dropout(0.5)(dense2)
    dense3 = Dense(256, activation='relu')(drop2)
    drop3  = Dropout(0.5)(dense3)
    out   = Dense(num_classes,\
                  kernel_regularizer=regularizers.l2(0.01),\
                  activity_regularizer=regularizers.l1(0.01),\
                  activation='softmax')(drop3)
    
    model = Model(input, out)
    
    model.summary()
    
    sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
    model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

    return model

model = get_model(BOW_LEN, CLASSES_NO)

history = model.fit(X_train, y_train,
                    batch_size=BATCH_SIZE,
                    epochs=EPOCHS,
                    verbose=1,
                    validation_split=0.1,
                    callbacks=[EarlyStopping(patience=2, monitor='val_loss')])

plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')   
plt.legend()
plt.show()
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
